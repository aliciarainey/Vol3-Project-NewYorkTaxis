```python
import numpy as np 
import pandas as pd
import pyarrow.parquet as pq
```

```python
tripslist2024 = [0] * 12
for i in range(1, 13):
    if i <= 9:
        trips = pq.read_table(f'yellow_tripdata_2024-0{i}.parquet')
        tripslist2024[i-1] = trips.to_pandas()
    else:
        trips= pq.read_table(f'yellow_tripdata_2024-{i}.parquet')
        tripslist2024[i-1] = trips.to_pandas()
    
trips2024 = pd.concat(tripslist2024, ignore_index=True)
```

```python
print(trips2024.isna().sum())
print(trips2024.shape)
#all the null entries appear to be missing from the same categories in the same row 
# if we do na.drop then shape of the array changes by the amount na in one column
#ask if this is okay
trips2024_cp = trips2024.dropna()
```

```python
print(trips2024_cp.columns)
print("pickup", trips2024_cp['tpep_pickup_datetime'].min(), trips2024_cp['tpep_pickup_datetime'].max())
print("dropoff", trips2024_cp['tpep_dropoff_datetime'].min(), trips2024_cp['tpep_dropoff_datetime'].max())
mask_pickup = (trips2024_cp['tpep_pickup_datetime'] > pd.Timestamp("2024-01-01 00:00:00")) & (trips2024_cp['tpep_pickup_datetime'] < pd.Timestamp("2025-01-01 00:00:00"))
trips2024_cp = trips2024_cp[mask_pickup]
mask_dropoff = (trips2024_cp['tpep_pickup_datetime'] > pd.Timestamp("2024-01-01 00:00:00")) & (trips2024_cp['tpep_pickup_datetime'] < pd.Timestamp("2025-01-01 00:00:00"))
trips2024_cp = trips2024_cp[mask_dropoff]
print("pickup", trips2024_cp['tpep_pickup_datetime'].min(), trips2024_cp['tpep_pickup_datetime'].max())
print("dropoff", trips2024_cp['tpep_dropoff_datetime'].min(), trips2024_cp['tpep_dropoff_datetime'].max())
```
